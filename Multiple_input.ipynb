```python
from typing import Dict, TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentInput(TypedDict):
    name: str
    age: int
    hobbies: List[str]
    
```

```python
def process_agent_input(state: AgentInput) -> AgentInput:

    state['name'] = "Hello, This is " + state['name'] + " and I am " + str(state['age']) + " years old. My hobbies are " + ", ".join(state['hobbies']) + "."
    return state
```

```python
graph = StateGraph(AgentInput)
graph.add_node("Node", process_agent_input)

graph.set_entry_point("Node")
graph.set_finish_point("Node")

app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
answer = app.invoke({"name": " Alice ", "age": 25, "hobbies": ["reading", "swimming", "cycling"]})
answer[ 'name']
```
