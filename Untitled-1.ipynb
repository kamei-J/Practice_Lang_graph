```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class MyState(TypedDict):
    name: str
    name1: str
    
def greet_agent_node(state: MyState) -> MyState:
    
    state['name'] = "Hey " + state["name"] + ", how are you doing today?"
    return state
```

```python
def farewall_agent_node(state: MyState) -> MyState:
    
    state['name1'] = "Goodbye " + state["name1"] + ", have a nice day!"
    return state
```

```python
graph = StateGraph(MyState)

graph.add_node("Gretting", greet_agent_node)
graph.add_node("Farewall", farewall_agent_node)

graph.set_entry_point("Gretting")
graph.add_edge("Gretting", "Farewall")
graph.set_finish_point("Farewall")

app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"name": " Alice ", "name1": " Alice "})

result["name"], result["name1"]
```
